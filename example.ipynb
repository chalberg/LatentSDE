## Querying MIMIC-IV in Python
Once you have access to the MIMIC-IV dataset in Google BigQuery, you are ready to start querying in Python. Pandas gbq is a package which allows us to query the MIMIC-IV dataset hosted on Google BigQuery and store the result in a pandas dataframe. The following code is a toy example of how to write and store the results of queries.

```python
from pandas.io import gbq

project = "mimic-380019" # replace with your Google Cloud project ID

# query first 10 rows of the vitalsign table
query = """
    SELECT *
    FROM physionet-data.mimiciv_derived.vitalsign
    LIMIT 100
    """

data = gbq.read_gbq(query, project_id=project) # this will prompt you to provide the account info associated w 

data.head(10)
```
